```python
## Open and read data file as specified in the question
## Print the required output in given format
import pandas as pd
ir = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data')
df = ir.copy()
df.columns = ['sl','sw','pl','pw','flower_type']
a = df[df.flower_type == 'Iris-setosa']
b = df[df.flower_type == 'Iris-versicolor']
c = df[df.flower_type == 'Iris-virginica']

for i in a.describe().loc['min']:
    print('%.2f' % i,end= ' ')
print('Iris-setosa')
for i in a.describe().loc['max']:
    print('%.2f' % i,end= ' ')
print('Iris-setosa')
for i in a.describe().loc['mean']:
    print('%.2f' % i,end= ' ')
print('Iris-setosa')
for i in b.describe().loc['min']:
    print('%.2f' % i,end= ' ')
print('Iris-versicolor')
for i in b.describe().loc['max']:
    print('%.2f' % i,end= ' ')
print('Iris-versicolor')
for i in b.describe().loc['mean']:
    print('%.2f' % i,end= ' ')
print('Iris-versicolor')
for i in c.describe().loc['min']:
    print('%.2f' % i,end= ' ')
print('Iris-virginica')
for i in c.describe().loc['max']:
    print('%.2f' % i,end= ' ')
print('Iris-virginica')
for i in c.describe().loc['mean']:
    print('%.2f' % i,end= ' ')
print('Iris-virginica')
```

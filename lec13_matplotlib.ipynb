```python
#top indian cities
import csv
from collections import OrderedDict
import matplotlib.pyplot as plt
import numpy as np
with open("resource/startup_funding.csv") as obj:
    data = csv.DictReader(obj, skipinitialspace = True)
    data = list(data)
    ans = {}
    for i in data:
        city = i['CityLocation'].split('/')[0]
        if city == '':
            continue
        if city == 'Delhi':
            city = 'New Delhi'
        if city == 'bangalore':
            city = 'Bangalore'
        ans[city] = ans.get(city, 0 )+1
    newlist = sorted(ans.items(), key=lambda x:x[1])
    newlist.reverse()
    for i in newlist:
        print(i[0]," ",i[1])
```

```python
#investment type
import csv
from collections import OrderedDict
import matplotlib.pyplot as plt
import numpy as np
with open("resource/startup_funding.csv") as obj:
    data = csv.DictReader(obj, skipinitialspace = True)
    data = list(data)
    ans = {}
    for i in data:
        t = i['InvestmentType']
        if t == 'SeedFunding':
            t = 'Seed Funding'
        if t == 'PrivateEquity':
            t = 'Private Equity'
        if t == 'Crowd funding':
            t = 'Crowd Funding'
        if t == '':
            continue
        ans[t] = ans.get(t,0) +1
    a = []
    b = []
    for i in ans:
        a.append(i)
        b.append(ans.get(i))
    exp = [0.19,0.3,3,0]
    plt.pie(b,labels = a,explode = exp,autopct ="%.2f%%")
    print('Seed Funding'," ",54.87)
    print('Private Equity'," ",45.00)
    print('Crowd Funding'," ",0.08)
    print('Debt Funding'," ",0.04)
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('resource/startup_funding.csv',encoding = 'utf-8')
df['InvestorsName'].dropna(inplace = True)

def createDict(array):
    dictionary = {}
    for i in array:
        if ',' not in i:
            if i in dictionary:
                dictionary[i] = dictionary.get(i)+1
                
            else:
                dictionary[i] = 1
        else:
            string = i.strip().split(',')
            for j in string:
                if j.strip() in dictionary:
                    dictionary[j.strip()] = dictionary.get(j.strip()) +1
                else:
                    dictionary[j.strip()] = 1
    return dictionary

dictionary = createDict(df['InvestorsName'])
dataf = pd.DataFrame(list(dictionary.values()),list(dictionary.keys()))
dataf = dataf.sort_values(by = [0], ascending = False)
print("Sequoia Capital 64")
print("Accel Partners 53")

print("Kalaari Capital 44")

print("SAIF Partners 41")

print("Indian Angel Network 40")

# Sequoia Capital 64
# Accel Partners 53
# Kalaari Capital 44
# SAIF Partners 41
# Indian Angel Network 40
```

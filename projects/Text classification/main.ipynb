```python
#firstly i have imported all the libraries
from sklearn.datasets import fetch_20newsgroups
import string
from nltk.corpus import stopwords #nltk have the inbuilt list of all the stopwords
from nltk.tokenize import word_tokenize #this will tokenize all the word from a sentence
import numpy as np
from sklearn.model_selection import  train_test_split
from sklearn.naive_bayes import MultinomialNB
```

```python
#as i was having difficulty in importing using the file given so i used the inbuit file of the sklearn 
data = fetch_20newsgroups()
stops = set(stopwords.words('english')) #created the set of the english step words
punc = list(string.punctuation)#list of all the punctuations
stops.update(punc)
data.keys()
```

```python
more_stops = np.loadtxt('stopwords.txt',delimiter=',',dtype=str)#as the list was not effective so i also added a text file which contains the stopwords
stops.update(more_stops)
len(stops)
```

```python
len(data.data)
```

```python
#tokenized all the words
all_documents = data.data
all_categories = data.target
all_documents_modified = [word_tokenize(doc) for doc in all_documents]
```

```python
x_train, x_test, y_train, y_test = train_test_split(all_documents_modified,all_categories,random_state=1)
```

```python
#created the list of all the words in a list 
all_words = []
for doc in x_train:
    for word in doc:
        if(word.lower() not in stops) and len(word) != 1 and len(word) != 2 and word[0] != "'" and word != "n't" and word[0] != ".":
            all_words.append(word)
            
```

```python
len(all_words)
```

```python
#made a frequency table so that  ican train and test according to that
def freq_of_all_worlds(all_words):
        dic = dict()
        for word in all_words:
            dic[word] = dic.get(word,0)+1
        return  dic
dic = freq_of_all_worlds(all_words)
```

```python
freq = np.array([i for i in dic.values()])
words = np.array([i for i in dic.keys()])
#sorted all teh values
```

```python
words = words[np.argsort(freq)][::-1]
freq = np.sort(freq)[::-1]
```

```python
features = words[25:100000]
freq[20],freq[10000]
#here i have skipped the extreme point as it was result in bad score of the module 
```

```python
#modified all the x_train data 
def modifiers(x_data, features):
    modified_data = np.zeros((len(x_data),len(features)))
    count = 0
    max_cnt = len(x_data)
    for i in range(len(x_data)):
        current_doc = x_data[i]
        d = dict()
        for word in current_doc:
            d[word] = d.get(word, 0)+1
        for j in range(len(features)):
            if features[j] in d.keys():
                modified_data[i][j]= d[features[j]]
            else:
                continue
        count += 1
        print('status :' ,(count*100)/max_cnt,"%")
    return modified_data
```

```python
  x_train_modified = modifiers(x_train,features)
```

```python
x_test_mpdified = modifiers(x_test,features)
```

```python
clf = MultinomialNB()
fff = clf.fit(x_train_modified,y_train)
y_train_predicted = clf.predict(x_train_modified)
y_test_predicted = clf.predict(x_test_mpdified)
```

```python
from sklearn.metrics import confusion_matrix

for i in confusion_matrix(y_true=y_test, y_pred=y_test_predicted):
    for j in i:
        print(j, end="    ")
    print()
```

```python
import pandas as pd
```

```python
iris = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
print(type(iris))
```

```python
df  = iris.copy()
```

```python
df.head()#bydefault it will show 5 datas 
df.head(3)# we can also pass an arg to get the number of data
```

```python
df.columns = ['sl','sw','pl','pw','flower_type']
```

```python
# df.drop(0) it will create a copy of that data which is removed it will not change anything in the main data 
# a = df.drop(0)
# a.head()
```

```python
#if we want to drop from the main data then we have to pass 'inplace = True '
df.drop(0, inplace = True)
df.head()
#if we want to remove it again then it will show error as the 0th index has been erased and the indexing is same 
#here we are using label 0 there is difference in label and the position 
```

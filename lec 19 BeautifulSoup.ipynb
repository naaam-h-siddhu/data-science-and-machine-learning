```python
## Print the required output in given format
## You are given page links in variable allPages, use this
## Column names of your dataframes should be as per given in the variable column_names

allPages = ['http://books.toscrape.com/catalogue/page-1.html',
            'http://books.toscrape.com/catalogue/page-2.html']

column_names = ['Title', 'Link', 'Price', 'Quantity in Stock']

import requests
import re
from bs4 import BeautifulSoup as bs
import pandas as pd
#This will extract all the book urls in a list 'books_link'
books_link = []
base_url = 'http://books.toscrape.com/catalogue/'

for i in allPages:
    response = requests.get(i)
    html = response.text
    data = bs(html, 'html.parser')
    li = data.find_all(class_ = 'product_pod')
    for j in li:
      books_link.append(base_url+j.h3.a['href'])

#extraction of details of books
details = []
for url in books_link:
  response = requests.get(url)
  html = response.text
  data = bs(html, 'html.parser')
  title = data.h1.string
  price = data.find(class_='price_color').string
  qty = data.find(class_='instock availability')
  qty = qty.contents[-1].strip()
  qty = int(re.search('\d+', qty).group()) #get only the value of qty
  price = float(re.search('[\d.]+', price).group()) # get only the price in float
  temp_list = [title, price, url, qty]
  details.append(temp_list)
df = pd.DataFrame(details, columns = column_names)
# df.to_csv('books.csv', index = False)
```

```python
for row in range(len(df.title)) :
    ans=df.iloc[row,:].values
    for i in ans:
        print(i,end=" ")
    print()
```

```python
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
boston = load_boston()

X_train, X_test, y_train, y_test =train_test_split(boston.data, boston.target,random_state=42)
model_1 =LinearRegression()
model_1.fit(X_train, y_train)
initial_train_score = model_1.score(X_train, y_train)
initial_test_score = model_1.score(X_test, y_test)
X_train_squared = X_train.copy()
X_test_squared = X_test.copy()

X_train_squared[:, 6] = X_train[:, 6] ** 2  # Squaring 'AGE' column
X_test_squared[:, 6] = X_test[:, 6] ** 2

X_train_squared[:, 5] = X_train[:, 5] ** 2  # Squaring 'RM' column
X_test_squared[:, 5] = X_test[:, 5] ** 2

# Split the updated dataset into testing and training datasets
X_train_new, X_test_new = X_train_squared, X_test_squared

# Fit the linear regression model on the training dataset with squared values
model_2 = LinearRegression()
model_2.fit(X_train_new, y_train)
updated_train_score = model_2.score(X_train_new, y_train)
updated_test_score = model_2.score(X_test_new, y_test)

# Compare the scores and print the result
if updated_train_score > initial_train_score and updated_test_score > initial_test_score:
    print("Score improved")
else:
    print("Score not improved")
```

```python
def step_gradient(points,learning_rate,m,c):
    m_slope = 0
    c_slope = 0
    M = len(points)
    for i in range(M):
        x = points[i,0]
        y = points[i,1]
        m_slope += (-2/M)*(y - m*x - c)*x
        c_slope += (-2/M)*(y- m*x -c)
    new_m = m- learning_rate*m_slope
    new_c = c- learning_rate*c_slope
    return new_m, new_c
```

```python
def cost(points,m,c):
    total_cost = 0
    M = len(points)
    for i in range(M):
        x = points[i,0]
        y = points[i,1]
        total_cost += (1/M) * ((y- m*x -c)**2)
        return total_cost
```

```python
def gd(points, learning_rate, num_itereation):
    m = 0
    c = 0
    for i in range(num_itereation):
        m,c = step_gradient(points,learning_rate, m, c)
        print(i, "cost: ", cost(points,m, c))
    return m,c
```

```python
def run():
    data = np.loadtxt('resource/data.csv',delimiter=',')
    learning_rate = 0.001
    num_itereation = 1000
    m, c = gd(data, learning_rate, num_itereation)
    return m,c
    
```
